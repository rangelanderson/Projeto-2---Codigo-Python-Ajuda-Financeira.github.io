<h2>Sobre o projeto</h2>
<p>Este código em Python visa transformar tabelas Excel de gastos mensais e anuais em gráficos visuais para melhor visualizção e compreensão. Facilitando a comparação do meu histórico de gastos e deste modo, me auxiliando no progresso da minha vida financeira.<br>
Obs: será feito passo a passo o código para auxiliar a compreensão.</p>

```python
# Passo 1: importar a tabela.

import pandas as pd

tabela = pd.read_excel("Gastos Totais Gerais 2026-2028.xlsx")
display(tabela)
```

```python
# Passo 2: Tratar e preparar a tabela para o gráfico. Transformando as palavras em números.

from sklearn.preprocessing import LabelEncoder

# coluna Despesas Fixas
print("Valores únicos na coluna Mês antes do LabelEncoder:")
print(tabela["Mês"].unique())
print("Tipo da coluna Mês:", tabela["Mês"].dtype)

codificador_mes = LabelEncoder()
tabela["Mês_Encoded"] = codificador_mes.fit_transform(tabela["Mês"])

print("Valores únicos após o LabelEncoder:")
print(tabela["Mês_Encoded"].unique())

display(tabela)
```

```python
# Passo 3: Melhorar visualização usando o plotly.

import plotly.express as px
import plotly.graph_objects as go
# Criando o gráfico de barras
grafico_um = px.histogram(data_frame=tabela, x="Gasto Total", y="Mês", text_auto=True)
grafico_um.show()

grafico_dois = px.histogram(data_frame=tabela, x="Ano", y="Gasto Total", text_auto=True)
grafico_dois.show()

grafico_tres = px.histogram(data_frame=tabela, x="Mês", y="Gasto Total", color="Ano", text_auto=True)
grafico_tres.show()

#Gráfico de Pizza 1
labels = tabela["Ano"]
values = tabela["Gasto Total"]

# Criando o gráfico de pizza
fig_um = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='value')])

# Mostrando o gráfico
fig_um.show()

# Gráfico de Pizza 2
labels = tabela["Ano"]
values = tabela["Gastos Fixos"]

# Criando o gráfico de pizza
fig_dois = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='value')])

# Mostrando o gráfico
fig_dois.show()

# Gráfico de Pizza 3
labels = tabela["Ano"]
values = tabela["Gastos Variáveis"]

# Criando o gráfico de pizza
fig_tres = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='value')])

# Mostrando o gráfico
fig_tres.show()
```
